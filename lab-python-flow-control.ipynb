```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}


print("Inventory:", inventory)
for product in products:
    quantity = int(input(f"Enter the quantity for {product}:"))
    inventory[product] = quantity    
    inventory_total = sum(inventory.values()) 
    print(inventory) 
print("Inventory:", inventory)


print(inventory_total)

#2 Prompt the user to enter the name of a product

customer_orders = set()                             

while len(customer_orders) < 3:
    order = input("Enter a product to order:")
    if order in products:
        customer_orders.add(order)                                              #a add the product name to the customer_orders set
        inventory_total -= 1
        inventory[order] -= 1
    else:
        print("Invalid product. Please choose from the available products.")
print("Customer orders:", customer_orders)

total_products_ordered = len(customer_orders)




print("Order statistics:")
print("Total products ordered:", total_products_ordered)
percentage_ordered = (total_products_ordered / inventory_total) * 100
print("Percentage of products ordered:", f"{percentage_ordered:.2f}%")

for item, qty in inventory.items():
    print(f"{item}: {qty} remaining")
```
